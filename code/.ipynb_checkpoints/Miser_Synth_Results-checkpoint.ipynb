```python
import matplotlib.pyplot as plt
from sklearn.datasets import dump_svmlight_file
from Greedymiser.miser import miser
import saeaborn as sns
sns.set(style="whitegrid")
```

```python
def generate_iid_reg(samples=1000,features=10,noise=0,weights=None,seed=100):
    np.random.seed(seed)
    X=np.random.rand(samples,features+noise)*100
    if weights is  None:
        weights=np.exp(-np.arange(1,features+1))
        weights=np.concatenate((weights/np.sum(weights),[0 for c in range(noise)]))
    y= np.sum(weights*X,axis=1)
    return X,y,weights

X,y,weights=generate_iid_reg(samples=5000,features=10,weights=None)
Y_train,Y_test=y[:2000],y[2000:]
X_train,X_test=X[:2000],X[2000:]
```

```python
dump_svmlight_file(X_test, Y_test, '../data/synth/svm/test.txt', zero_based=False, comment=None, query_id=[0]*len(Y_test), multilabel=False)
dump_svmlight_file(X_train, Y_train, '../data/synth/svm/train.txt', zero_based=False, comment=None, query_id=[0]*len(Y_train), multilabel=False)
```

```python
steps=[0.01,0.1,0.5,0.9]
sets=[]
for seed in range(20):
    X,y,weights=generate_iid_reg(samples=5000,features=10,weights=None,seed=seed)
    Y_train,Y_test=y[:2000],y[2000:]
    X_train,X_test=X[:2000],X[2000:]
    dump_svmlight_file(X_test, Y_test, '../data/synth/svm/test.txt', zero_based=False, comment=None, query_id=[0]*len(Y_test), multilabel=False)
    dump_svmlight_file(X_train, Y_train, '../data/synth/svm/train.txt', zero_based=False, comment=None, query_id=[0]*len(Y_train), multilabel=False)
    
    results=[]
    for step in steps:
        x=miser()
        x.DEPTH=10
        x.STEPSIZE=step
        x.ITER=100
        report=x.fit(path='synth',costs= "s",l =5)
        results.append(report)
        print("done step: %f" % step)
    sets.append(results)
    print("done run %i",seed)
sets=np.array(sets)
```

```python
plt.figure(figsize=(20,5))
for s in range(len(steps)):
#     plt.errorbar(range(100),sets[:,s,:,2].mean(axis=0),yerr=sets[:,s,:,2].std(axis=0,dtype=np.float64),label=steps[s])

    violin_parts=plt.violinplot(sets[:,s,:,2].mean(axis=0),
                   positions=np.arange(x.ITER),showmeans=True,widths=0.8)
    c=tuple(violin_parts['bodies'][0].get_facecolors()[0][:-1])


    plt.plot(range(x.ITER),sets[:,s,:,2].mean(axis=0),
             c=c,label=steps[s])    
    
plt.legend(loc="upper right",title="Step size")
plt.title("Step size and number of iterations influence on performance")
plt.xlabel("Number of iterations")
plt.ylabel("Test RMSE")
plt.show()
```
